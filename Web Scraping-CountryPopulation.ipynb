```python
# import the library we use to open URLs
import urllib.request
```

```python
# specify which URL/web page we are going to be scraping
url = "https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)"
```

```python
# open the url using urllib.request and put the HTML into the page variable
page = urllib.request.urlopen(url)
```

```python
# import the BeautifulSoup library so we can parse HTML and XML documents
from bs4 import BeautifulSoup
```

```python
# parse the HTML from our URL into the BeautifulSoup parse tree format
soup = BeautifulSoup(page, "lxml")
```

```python
print(soup.prettify())
```

```python
all_tables=soup.find_all("table")
all_tables[3]
```

```python
right_table=soup.find('table', id='main')
right_table
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
#G=[]
#H=[]
#I=[]
#J=[]
#K=[]
#L=[]


for row in all_tables[3].findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==6:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
        D.append(cells[3].find(text=True))
        E.append(cells[4].find(text=True))
        F.append(cells[5].find(text=True))
#        G.append(cells[6].find(text=True))
#        H.append(cells[7].find(text=True))
#        I.append(cells[8].find(text=True))
#        J.append(cells[9].find(text=True))
#        K.append(cells[10].find(text=True))
#        L.append(cells[11].find(text=True))
```

```python
import pandas as pd
df=pd.DataFrame(A,columns=['Country'])
df['Region']=B
df['Subregion']=C
df['Population 2018']=D
df['Population 2019']=E
df['Change']=F
# df['']=G
# df[' ']=H
# df[' ']=I
# df[' ']=J
# df[' ']=K
# df[' ']=L
df
```
